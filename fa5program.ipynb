```python
def get_choice():
    while True:
        choice = input("Choose an option:\n1) Calculate total cost\n2) Exit\nEnter 1 or 2: ")
        if choice in ("1", "2"):
            return choice
        print("Please enter 1 or 2.")


def get_price():
    while True:
        try:
            return float(input("what price of one item "))
        except ValueError:
            print("Invalid input. Please enter a numeric value for the price.")


def get_quantity():
    while True:
        try:
            return int(input("how many items "))
        except ValueError:
            print("Invalid input. Please enter an integer value for the quantity.")


def calculate_total(price, quantity):
    return price * quantity


def main():
    choice = get_choice()
    if choice == "1":
        price = get_price()
        quantity = get_quantity()
        total = calculate_total(price, quantity)
        print(f"Total cost: {total}")
    else:
        print("Goodbye!")


if __name__ == "__main__":
    main()
```
